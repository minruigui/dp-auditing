```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist

# Load the data
(trn_x, trn_y), (tst_x, tst_y) = fashion_mnist.load_data()

# Define the class names for the labels
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

# Randomly select 9 indices from the training set
indices = np.random.choice(trn_x.shape[0], 9, replace=False)
print(trn_x.shape)
plt.figure(figsize=(10, 10))
for i, idx in enumerate(indices):
    plt.subplot(3, 3, i+1)
    plt.imshow(trn_x[idx], cmap='gray')
    plt.title(class_names[trn_y[idx]])
    plt.axis('off')
plt.show()
```
